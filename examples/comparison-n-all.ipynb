```python
import algorithms
import treespace
import logging 
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def calculate_l(n):
    x = 0
    for i in range(n-1, -1, -2):
        x += 2**i
    return x
```

```python
logging.basicConfig(filename='logs.txt', format='%(levelname)s:%(message)s', level=logging.DEBUG)
```

```python
results = []
results_df = pd.DataFrame(columns = ["n", "k", "len", "alg", "tree_caching"])
repeats_per_k = 1
repeats = 10
for n in range(4, 10):
    print("n =", n)
    for repeat in range(repeats):
        print("-- repeat =", repeat)
        tree = treespace.BinaryTree(n)
        alg = algorithms.RandomBucket(tree)
        tree_caching = algorithms.TreeCaching(tree)
        requests_idx = random.zipf(a=1.5, size=1000)
        requests_idx = list(2**n - requests_idx[requests_idx <= tree.get_size()])
        req_len = len(requests_idx)
        requests = list(map(lambda x: tree.nodes[x - 1], requests_idx))

        for k in range(4, 2**(n-1), 1):
            print("---- k =", k)
            for repeat_same_k in range(repeats_per_k):
                print("------ repeats_same =", repeat_same_k)
                print(f"-------- Request Len = {req_len}")
                alg = algorithms.RandomBucket(tree)
                alg_cost = alg.serve_requests(requests, k, calculate_l(n), 1)
                print(f"-------- Alg Cost = {alg_cost}")
                tree_c_cost = tree_caching.serve_requests(requests, k, 1)
                print(f"-------- Tree C Cost = {tree_c_cost}")
                results.append({'n':n, 'k':k, 'len': req_len, 'alg':alg_cost, 'tree_caching':tree_c_cost})
                results_df = results_df.append({'n':n, 'k':k, 'len': req_len, 'alg':alg_cost, 'tree_caching':tree_c_cost}, 
                                 ignore_index=True)
```

```python
results_df
```

```python
results_df['alg_ratio'] = results_df['alg']/results_df['len']
results_df['tc_ratio'] = results_df['tree_caching']/results_df['len']
results_df
```

```python
results_arr = results_df[['n','k','alg_ratio','tc_ratio']].values.copy().astype("float")
res_smooth = np.array([[]]).reshape((0,4))
for n in range(4,10):
    for k in range(4, 2**(n-1), 1):
        res = np.mean(results_arr[(results_arr[:,0] == n) & (results_arr[:,1] == k)], axis=0).reshape((1,4))
        res_smooth = np.append(res_smooth, res, axis=0)
```

```python
# n = 4
n = 4
data = res_smooth[res_smooth[:, 0]==n][:, [1,2,3]]
plt.figure(figsize=(10,7))
plt.plot(data[:,0], data[:,1], 'r', linewidth=1.5)
plt.plot(data[:,0], data[:,2], 'b', linewidth=1.5)
plt.xlabel("Cache Size")
plt.ylabel("Cost/(Sequence Length)")
plt.title(f"Cost/(Sequnce Length) vs Cache Size (For n = {n})")
plt.grid()
plt.legend(["Random Bucket with Bypassing", "Tree Caching"])
plt.savefig(f"cost_cache_n={n}.png", format='png', dpi=300, bbox_inches='tight')
```
